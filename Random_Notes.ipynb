Functions returning multiple values

```python
def return_powers(n):
    square = n**2
    cube = n**3
    four = n**4
    return square,cube,four
```

```python
print(return_powers(4))
```

```python
a = return_powers(4)
```

```python
print(a)
```

```python
type(a)
```

In case of list, if you define another variable as a refernce, both of them point to the same memory location. So if you change list element, using the reference variable, you will see the same change reflected while referring to the list using the original variable name. See example below

```python
a = [1,2,3]
b=a
b[0] = 10
print(a)
```

This does not happen in case of regular variables

```python
a = 1
b = a
# Uptil this statement, both a and b is pointing to same memory location
# Now I try to change the value of b
b = 2
# Now b is no longer pointing to the memory location storing the value 1. It is not pointing to a new memory location 
# storing the value of 2
# This is different from how list behaves
# In case of list, both the original variable and new reference variable keeps pointing to the same memory (Call by reference)
```
